```python
from function import *

with open(path_dictionary, "r") as a:
    name_dict = json.load(a)
```

```python
stock_data = clean_daily_data(path_stock_data, path_ticker, name_dict)
```

```python
market = (
    stock_data.groupby(["exchange", "time"])
    .agg(
        market_return=("return_weighted", "sum"),
        market_log_return=("log_return_weighted", "sum"),
    )
    .reset_index()
)
```

```python
df = stock_data.merge(market, how="left", on=["time", "exchange"])
```
